```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

1.Create two 3Ã—3 matrices with random integers in Python.

```python
matrix_1 = np.random.randint(1, 10, (3, 3))
matrix_2 = np.random.randint(1, 10, (3, 3))

print("Matrix 1:")
print(matrix_1)
print("\nMatrix 2:")
print(matrix_2)
```

2. Perform addition, subtraction, and multiplication operations on these matrices.


```python
# Addition
matrix_sum = np.add(matrix_1, matrix_2)
print("\nMatrix Sum:")
print(matrix_sum)

# Subtraction
matrix_difference = np.subtract(matrix_1, matrix_2)
print("\nMatrix Difference:")
print(matrix_difference)

# Multiplication
matrix_product = np.dot(matrix_1, matrix_2)
print("\nMatrix Product:")
print(matrix_product)
```

3. Compute the determinant, inverse, and rank of one of the matrices.

```python
# Determinant of Matrix 1
determinant_matrix1 = np.linalg.det(matrix_1)
print("\nDeterminant of Matrix 1:")
print(determinant_matrix1)

# Inverse of Matrix 1 using if-else
if not np.isclose(determinant_matrix1, 0): # Check if determinant is not close to zero
    inverse_matrix1 = np.linalg.inv(matrix_1)
    print("\nInverse of Matrix 1:")
    print(inverse_matrix1)
else:
    print("\nMatrix 1 is singular (determinant is zero), so its inverse cannot be computed.")

# Rank of Matrix 1
rank_matrix1 = np.linalg.matrix_rank(matrix_1)
print("\nRank of Matrix 1:")
print(rank_matrix1)
```
