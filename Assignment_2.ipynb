```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

**1. Generate and Plot Two Random Vectors:**

a. Create two vectors, each containing 15 random floats of your choice.


```python
vector_1 = np.random.rand(15)
vector_2 = np.random.rand(15)

print("Vector 1:", vector_1)
print("Vector 2:", vector_2)
```

b. Plot these vectors on the same graph using matplotlib to compare how
they look.
c. Label the axes appropriately and give the plot a title.

```python
plt.figure(figsize=(10, 6))
plt.plot(vector_1, label='Vector 1')
plt.plot(vector_2, label='Vector 2')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Two Random Vectors')
plt.legend()
plt.grid(True)
plt.show()
```

**2. Generate and Visualize a Random Matrix:**


a. Create a 4x4 matrix with random values of your choice.


```python
matrix = np.random.rand(4, 4)
print("Random 4x4 Matrix:\n", matrix)
```

b. Display the matrix using a heatmap with appropriate color mapping (e.g.,
seaborn or matplotlib).
c. Add labels to the heatmap to indicate row and column numbers.

```python
plt.figure(figsize=(6, 5))
sns.heatmap(matrix, annot=True, cmap='viridis', fmt=".2f",
            xticklabels=np.arange(1, matrix.shape[1] + 1),
            yticklabels=np.arange(1, matrix.shape[0] + 1))
plt.xlabel('Column Number')
plt.ylabel('Row Number')
plt.title('Heatmap of Random 4x4 Matrix')
plt.show()
```

**3. Matrix Operations and Visualization:**


a. Create two 4x4 matrices with random values of your choice.


```python
matrix_a = np.random.rand(4, 4)
matrix_b = np.random.rand(4, 4)

print("Matrix A:\n", matrix_a)
print("\nMatrix B:\n", matrix_b)
```

b. Perform the following operations:
i. Matrix Addition
ii. Matrix Subtraction
iii. Matrix Multiplication (ensure matrices are compatible for
multiplication)



```python
matrix_addition = matrix_a + matrix_b
matrix_subtract = matrix_a - matrix_b
matrix_multiply = np.dot(matrix_a, matrix_b)

print("Matrix Addition (A + B):\n", matrix_addition)
print("\nMatrix Subtraction (A - B):\n", matrix_subtract)
print("\nMatrix Multiplication (A * B):\n", matrix_multiply)
```

c. For each of these operations, visualize the result using bar plots. Each bar
should represent one element of the matrix.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot Matrix Addition
sns.barplot(x=np.arange(matrix_addition.size), y=matrix_addition.flatten(), ax=axes[0], color='skyblue')
axes[0].set_title('Matrix Addition (A + B)')
axes[0].set_xlabel('Element Index')
axes[0].set_ylabel('Value')

# Plot Matrix Subtraction
sns.barplot(x=np.arange(matrix_subtract.size), y=matrix_subtract.flatten(), ax=axes[1], color='lightcoral')
axes[1].set_title('Matrix Subtraction (A - B)')
axes[1].set_xlabel('Element Index')
axes[1].set_ylabel('Value')

# Plot Matrix Multiplication
sns.barplot(x=np.arange(matrix_multiply.size), y=matrix_multiply.flatten(), ax=axes[2], color='lightgreen')
axes[2].set_title('Matrix Multiplication (A * B)')
axes[2].set_xlabel('Element Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
